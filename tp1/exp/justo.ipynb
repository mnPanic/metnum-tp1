```python
import utils
import pandas as pd

def read_govan(path: str) -> pd.DataFrame:
    govan = utils.read_ranking(path)
    govan["team"] = (govan.index + 1)
    return govan.sort_values(["rating"], ascending=False).reset_index(drop=True)

DATA_DIR = "../data/"
GOVAN_ORIG = DATA_DIR + "govan.dat"
GOVAN_CH = DATA_DIR + "govan_changed.dat"
GOVAN_NEW = DATA_DIR + "govan_new.dat"
```

```python
utils.run(GOVAN_ORIG, utils.METHOD_CMM)
read_govan(GOVAN_ORIG+utils.METHOD_CMM+utils.OUT)
```

```python
utils.run(GOVAN_CH, utils.METHOD_CMM)
read_govan(GOVAN_CH+utils.METHOD_CMM+utils.OUT)
```

```python
utils.run(GOVAN_NEW, utils.METHOD_CMM)
read_govan(GOVAN_NEW+utils.METHOD_CMM+utils.OUT)
```
